```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```
df_mcd=pd.read_excel('https://github.com/insaid2018/Term-1/raw/master/Data/Casestudy/mcdonalds_india_menu.xlsx')
df_mcd.head()
```

```
df_store=pd.read_excel('https://github.com/insaid2018/Term-1/raw/master/Data/Casestudy/mcdonalds_store_data.xlsx')
df_store.head()
```

```
df_mcd["Category"].value_counts().plot.bar(figsize=(15,7),color="g",fontsize=13,yticks=np.arange(0,19,3))
plt.xlabel("categories")
plt.ylabel("counts")
plt.title("Ctegory counts")
```

```
(df_mcd["Category"].value_counts()/len(df_mcd)).plot.bar(figsize=(15,7),fontsize=13,color="r")
```

```
df_mcd.head()
```

```
df_mcd["Energy"].plot.line(figsize=(13,7),color="r",linestyle="-.")
```

```
df_mcd["Category"].iloc[15:25].value_counts()
```

```
df_mcd.head()
```

```
df_mcd["Total_Fat"].plot.area(figsize=(13,7),fontsize=13,color="g")
```

```
df_mcd["Serve_Size"].max()
```

```
df_mcd["Serve_Size"].min()
```

```
bins=[0,50,100,150,200,250,300,350,400,450,500]
df_mcd["Serve_Size"].plot.hist(bins=bins,figsize=(15,7),fontsize=13,color="b",xticks=bins)
```

```
df_mcd["Serve_Size"].plot.kde(figsize=(15,7),color="g")
```

```
space = np.ones(11)/10
df_mcd['Category'].value_counts().plot(kind='pie', explode=space, fontsize=14, autopct='%3.1f%%', wedgeprops=dict(width=0.15), 
                                       shadow=True, startangle=160, figsize=(10,10), cmap='inferno', legend=True)

plt.ylabel('Category')
plt.title('Donut Plot showing the proportion of each Category value')
```

```
df_mcd.plot.scatter(x="Carbohydrates",y="Energy")
```

```
plt.figure(figsize=(15,7))
sns.regplot(x="Carbohydrates",y="Energy",data=df_mcd,)
```

```
df_mcd.plot.hexbin(x="Carbohydrates",y="Energy",figsize=(15,7),gridsize=12)
```

```
df_mcd.head()
```

```
df_mcd.groupby(["Category"])["Total_Fat"].mean().plot.bar(figsize=(15,7),fontsize=13,color="g")
df_mcd.groupby(["Category"])["Saturated_Fat"].mean().plot.bar(figsize=(15,7),fontsize=13,color="r")
df_mcd.groupby(["Category"])["Trans_Fat"].mean().plot.bar(figsize=(15,7),fontsize=13,color="b")
plt.legend(["Total_Fat","Saturated_Fat","Trans_Fat"])
```

```
df_mcd.groupby(["Serve_Size"])["Energy"].mean().plot.line(figsize=(15,7),fontsize=13,color="r",linestyle="-.")
df_mcd.groupby(["Serve_Size"])["Cholestrol"].mean().plot.line(figsize=(15,7),fontsize=13,color="g",linestyle="-.")
plt.legend(["Energy","Cholestrol"])
df_mcd.info()
```

```
df_store.groupby(["Customers"])["Revenue"].mean().plot.area(figsize=(15,7),fontsize=13,color="b")
df_store.groupby(["Customers"])["Profits"].mean().plot.area(figsize=(15,7),fontsize=13,color="g")
plt.legend(["Revenue","Profits"])
```

```
sns.pairplot(data=df_mcd[["Energy","Total_Fat","Cholestrol","Protein","Sugars"]],size=2.5,diag_kind="kde")
```

```
plt.figure(figsize=(13,7))
sns.scatterplot(data=df_mcd,x="Energy",y="Sugars",hue="Category")
```

```
plt.figure(figsize=(13,7))
sns.scatterplot(data=df_mcd,x="Energy",y="Cholestrol",hue="Category")
```

```
plt.figure(figsize=(13,7))
sns.boxplot(data=df_store,x="State",y="Revenue",)
```

```
corr_mat=df_mcd.corr()
plt.figure(figsize=(12,8))
sns.heatmap(corr_mat,annot=True)
```

```
df_cp=df_mcd.copy()
```

```
df_cp.head()
```

```
plt.figure(figsize=(16,9))
parallel_coordinates(df_cp,class_column="High_Or_Low_Sugar",cols=["Energy","Protein","Total_Fat","Carbohydrates"])
```

```
plt.figure(figsize=(16,9))
parallel_coordinates(df_cp, 
                     class_column='High_Or_Low_Sugar', colormap='viridis',  
                     cols=['Energy', 'Protein', 'Total_Fat', 'Carbohydrates'])
```

```
plt.figure(figsize=(15,7))
sns.violinplot(data=df_mcd,y="Energy",x="High_Or_Low_Sugar")
```
